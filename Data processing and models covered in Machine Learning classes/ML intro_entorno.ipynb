## ¿Qué es pandas?

Es la librería más conocida para manipular datos en Python. El formato más utilizado es el de Dataframe y se podría decir que es el Excel de Python.

La principal característica de Pandas es su estructura de datos de alto nivel, que se conoce como DataFrame. Un DataFrame de Pandas es similar a una hoja de cálculo en un programa como Microsoft Excel, pero tiene capacidades y funcionalidades adicionales que lo hacen muy útil para el análisis de datos.

Cargamos los datos con pandas

```python
# LEO datos

datos_direccion = "Lionel-Messi-Club-Goals/data.csv"

df_messi = pd.read_csv(datos_direccion, encoding_errors="replace")

# muestro los datos
df_messi
```

```python
# fijemonos que en Minute hay algunos con el apóstrofe y otros que no
# quitemos ese apostrófe
df_messi["Minute"] = df_messi["Minute"].str.replace("'", "")

# mostremos los datos
df_messi
```

```python
# cuántos registros hay por club
df_messi["Club"].value_counts()
```

```python
# veamos en qué tiempos marcó más goles
df_messi["Minute"].value_counts()
```

```python
# primero veamos cuales registros hay con + (son los que agregan minutos)
df_messi[df_messi["Minute"].str.contains("\+")]
```

```python
# hagamos que simplemente sean los minutos sin agregar nada
# sería pasar de 90+2 a 90. quedarnos con lo que está a la izq del +
df_messi["Minute"] = df_messi["Minute"].str.split("\+").str[0]
```

```python
# contemos cuántos registros hay con + (ahora deberían ser 0)
df_messi[df_messi["Minute"].str.contains("\+")]
```

```python
# separemos en grupos de a 10 minutos. es decir, si Messi metió gol a
# los 14 minutos, diremos que lo metió en el grupo de los 10-20 minutos

# para eso, primero pasemos Minute a int
df_messi["Minute"] = df_messi["Minute"].astype(int)

# hacer grupos entre 0-9, 10-19, 20-29, etc
df_messi["Minute_decenas"] = (df_messi["Minute"] // 10) * 10

df_messi
```

```python
# ver cuántos goles metió en cada grupo de minutos decenas
df_messi["Minute_decenas"].value_counts().sort_index()
```

```python
# graficamos
df_messi["Minute_decenas"].value_counts().sort_index().plot(kind="bar",
                                                            figsize=(12, 10))
```

```python
# cuandó hizo más goles? en el primero o en el segundo tiempo?
# recordar que el primer tiempo es entre el minuto 0 y 45, y el segundo entre 46 y 90

# creamos la variable primer_tiempo
df_messi["primer_tiempo"] = df_messi["Minute"] <= 45
# creamos la variable segundo_tiempo
df_messi["segundo_tiempo"] = df_messi["Minute"] > 45

# mostramos rápido los datos
df_messi.head()
```

```python
# cuántos goles hizo en el primer tiempo
df_messi["primer_tiempo"].sum()
```

```python
# cuántos goles hizo en el segundo tiempo
df_messi["segundo_tiempo"].sum()
```

```python
# quién asistió más a Messi?
df_messi["Goal_assist"].value_counts()
```

```python
# grafiquemos los primeros 10
df_messi["Goal_assist"].value_counts().head(10).plot(kind="barh", figsize=(10, 6))
```

```python
# a qué oponente le hizo más goles?
df_messi["Opponent"].value_counts()
```

```python
# grafiquemos los primeros 20 en un gráfico de torta con el número de goles
df_messi["Opponent"].value_counts().head(20).plot(
    kind="pie", figsize=(10,10), autopct="%1.1f%%",
    fontsize=12, title="Equipos a los que Messi más goles hizo",
    ylabel=""
)
```
