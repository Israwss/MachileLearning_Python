8. "loan": tiene crditos de consumo? (categrica: "yes", "no")
9. "contact": medio a travs del cual fue contactado (categrica: "unknown", "telephone", "cellular")
10. "day": ltimo da del mes en el que fue contactada (numrica)
11. "month": ltimo mes en el que fue contactada (categrica: "jan", "feb", "mar", ..., "nov", "dec")

#4. Limpieza

Realizaremos el proceso de limpieza teniendo en cuenta las situaciones ms comunes:

1. Datos faltantes en algunas celdas
2. Columnas irrelevantes (que no responden al problema que queremos resolver)
3. Registros (filas) repetidos
4. Valores extremos (*outliers*) en el caso de las variables numricas. Se deben analizar en detalle pues no necesariamente la solucin es eliminarlos
5. Errores tipogrficos en el caso de las variables categricas



## 4.1 Datos faltantes

Ac comenzamos a ver que los datos no estn completos, pues no todas las columnas tienen la misma cantidad de registros.

El nmero total de registros debera ser 45.215. Sin embargo columnas como "job", "marital", "education", "balance", "duration" y "pdays".

Por ser tan pocos los datos  faltantes optamos por eliminar las filas correspondientes:

```python
data.dropna(inplace=True)
data.info()
```

##4.2 Columnas irrelevantes

Una columna irrelevante puede ser:

- **Una columna que no contiene informacin relevante para el problema que queremos resolver**. Por ejemplo en este caso podra ser una columna que no guarde relacin con el posible perfil del cliente (deporte favorito, hobbies, comida favorita, etc.)
- **Una columna categrica pero con un slo nivel**. Por ejemplo si en la columna "job" solo tuvisemos el nivel "unknown".
- **Una columna numrica pero con un slo valor**. Por ejemplo si en la columna "edad" todos los vlaores fuesen iguales a 50.
- **Columnas con informacin redundante**. Por ejemplo si adems de las columnas "month" y "day" tuvisemos la columna "month-day", resultado de combinar las dos anteriores.

Si tenemos la duda de si alguna columna puede ser relevante o no lo mejor es dejarla (y ms adelante en posteriores etapas podremos darnos cuenta de si se debe conservar o no).

En este caso todas las columnas pueden resultar relevantes, pero debemos verificar que no haya columnas categricas con un slo nivel, o columnas numricas con un slo valor:

```python
# Conteo de los niveles en las diferentes columnas categricas
cols_cat = ['job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'poutcome', 'y']

for col in cols_cat:
  print(f'Columna{col}: {data[col].nunique()} subniveles')
```

Todas las columnas categricas tienen ms de 1 subnivel. No eliminaremos ninguna.

Verifiquemos lo que ocurre con las columnas numricas:

```python
data.describe()
```

Todas las columnas numricas tienen desviaciones estndar ("std") diferentes de cero, lo que indica que no tienen un nico valor.

Preservaremos todas las columnas numricas.

##4.3 Filas repetidas

```python
print(f'Tamao del set antes de eliminar las filas repetidas: {data.shape}')
data.drop_duplicates(inplace=True)
print(f'Tamao del set despus de eliminar las filas repetidas: {data.shape}')
```

##4.4 *Outliers* en las variables numricas

No siempre se deben eliminar los *outliers* porque dependiendo de la variable numrica analizada estos pueden contener informacin importante.

Creemos grficas tipo "boxplot" de las columnas numricas:

```python
import seaborn as sns
```

```python
# Generar grficas individuales pues las variables numricas
# estn en rangos diferentes


cols_num = ['age', 'balance', 'day', 'duration', 'campaign',
            'pdays', 'previous']

fig, ax = plt.subplots(nrows=7, ncols=1, figsize=(8,30))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(cols_num):
    sns.boxplot(x=col, data=data, ax=ax[i])
    ax[i].set_title(col)
```

**Observaciones:**
- "age": hay sujetos con edades mucho mayores a 100 aos
- "duration": hay valores negativos
- "previous": hay un valor extremadamente alto (cercano a 300)

```python
# Eliminar filas con "age">100
print(f'Tamao del set antes de eliminar registros de edad: {data.shape}')
data = data[data['age']<=100]
print(f'Tamao del set despus de eliminar registros de edad: {data.shape}')
```

```python
# Eliminar filas con "duration"<0
print(f'Tamao del set antes de eliminar registros de duracin: {data.shape}')
data = data[data['duration']>0]
print(f'Tamao del set despus de eliminar registros de duracin: {data.shape}')
```

```python
# Eliminar filas con "previous">100
print(f'Tamao del set antes de eliminar registros de previous: {data.shape}')
data = data[data['previous']<=100]
print(f'Tamao del set despus de eliminar registros de previous: {data.shape}')
```

##4.5 Errores tipogrficos en variables categricas

En una variable categrica pueden aparecer sub-niveles como "unknown" y "UNK" que para nosotros son equivalentes pero que para nuestro analaisis pareceran diferentes.

Se deben unificar estos sub-niveles

```python
# Graficar los subniveles de cada variable categrica
cols_cat = ['job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'poutcome', 'y']

fig, ax = plt.subplots(nrows=10, ncols=1, figsize=(10,30))
fig.subplots_adjust(hspace=1)

for i, col in enumerate(cols_cat):
  sns.countplot(x=col, data=data, ax=ax[i])
  ax[i].set_title(col)
  ax[i].set_xticklabels(ax[i].get_xticklabels(),rotation=30)
```

Inicialmente se observa que hay sub-niveles con el mismo nombre pero escritos en minscula, en mayscula o con la primera letra en mayscula.

Unificamos estos sub-niveles:

```python
for column in data.columns:
    # Representar en minsculas slo si la columna es categrica
    if column in cols_cat:
        data[column] = data[column].str.lower()

# Y generemos las grficas nuevamente
fig, ax = plt.subplots(nrows=10, ncols=1, figsize=(10,30))
fig.subplots_adjust(hspace=1)

for i, col in enumerate(cols_cat):
  sns.countplot(x=col, data=data, ax=ax[i])
  ax[i].set_title(col)
  ax[i].set_xticklabels(ax[i].get_xticklabels(),rotation=30)
```

```python
# job: unificar admin. y administrative
print(data['job'].unique())
data['job'] = data['job'].str.replace('admin.','administrative', regex=False)
print(data['job'].unique())
```

```python
# marital: unificar div. y divorced
print(data['marital'].unique())
data['marital'] = data['marital'].str.replace('div.','divorced', regex=False)
print(data['marital'].unique())
```

```python
# education: unificar sec. y secondary, unk y unknown
print(data['education'].unique())
data['education'] = data['education'].str.replace('sec.','secondary', regex=False)
data.loc[data['education']=='unk','education'] = 'unknown'
print(data['education'].unique())
```

```python
# contact: unificar telephone y phone
print(data['contact'].unique())
data.loc[data['contact']=='phone','contact'] = 'telephone'
data.loc[data['contact']=='mobile','contact'] = 'cellular'
print(data['contact'].unique())
```

```python
# poutcome: unificar unk y unknown
print(data['poutcome'].unique())
data.loc[data['poutcome']=='unk','poutcome']='unknown'
print(data['poutcome'].unique())
```
